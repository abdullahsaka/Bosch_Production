```python
# import Timestamp data
def import_date(path):
    dfList = []
    df=pd.DataFrame()
    for iot_chunk in pd.read_csv(path,chunksize=100000):
        dfList.append(iot_chunk)
    df = pd.concat(dfList,sort=False)
    return df
```

# Case Study 3: Clustering Analysis | Explore production journey of 2019 parts

```python
pzs = pzs.set_index(['Id'])
pzs['L0_S0_D0']=0

cols = list(pzs.columns)
cols = [cols[-1]] + cols[:-1]
pzs = pzs[cols]

pzs=pzs.fillna(0) # convert NaN into 0
pzs.head(3)
```

```python
pzs=pzs.diff(axis=1) # lag difference
```

```python
q=pd.DataFrame(pzs.mask(pzs>=0).stack().index) # parts are going to which lines & stations
q[['Part', 'Flow']] = pd.DataFrame(q[0].tolist(), index=q.index) 
q.drop([0],axis=1,inplace=True)
q.head(3)
```

```python
Numofcheck=q.groupby('Part').count() # count checkpoints
Numofcheck=pd.DataFrame(Numofcheck).reset_index() # part - flow(lenght) dataframe
```

```python
np.sort(Numofcheck['Flow'].unique()) # cluster lenghts
```

```python
Numofcheck.nunique()['Flow'] # the number of clusters
```

```python
# build a dataframe including part, length of the journey,response
binary_suc={}
for j in Numofcheck.Part.unique():
    binary_suc[j]= numer[numer.Id==j]['Response'].item()

resp=pd.DataFrame.from_dict(binary_suc,orient='index',columns=['Response']).reset_index()
resp=resp.rename(columns={'index':'Part'})

dat_2019=pd.merge(Numofcheck,resp,how='left',on='Part')
dat_2019.head(3)
```

```python
# show the number of failed vs qualified poroducts for each cluster
a=dat_2019.groupby('Flow').sum()['Response']
b=dat_2019.groupby('Flow').count()['Response']
clus=pd.concat((a.rename('Fail'), b.rename('Total')),axis=1)
clus.head(3)
```
